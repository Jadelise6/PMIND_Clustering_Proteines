# Création du fichier csv contenant toutes les protéines pour créer les embeddings

```python
import os
```

### Obtention de la liste de tous les fichiers .fasta

```python
# Chemin du répertoire
path = "../PMIND2026_cluster/darkdino_fasta"

# Obtention de la liste de tous les fichiers et dossiers
fichiers = os.listdir(path)
```

```python
for fichier in fichiers:
    print(fichier)
```

### Lecture des fichiers et écriture de toutes les protéines dans all_dardino.csv

```python
all_darkdino_path = "./Data/all_darkdino.csv"

for fichier in fichiers:
    with open(f"{path}/{fichier}", 'r') as f:
        lignes = f.readlines()

    with open(all_darkdino_path, 'a') as f:
        for ligne in lignes:
            # Repérer les protéines = séquence d'acides aminés
            if ligne[0] != ">":
                f.write(ligne)
```
